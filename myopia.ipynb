```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Import the data
file = "Resources/myopia.csv"
df = pd.read_csv(file)
df.head()
```

Part 1

```python
# Create new dataframe by removing "Myopic" to make it unsupervised model
new_df = df.drop(["MYOPIC"], axis='columns')
new_df.head()
```

```python
# Find null values
for column in new_df.columns:
    print(f"Column {column} has {new_df[column].isnull().sum()} null values")
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(new_df[['AGE', 'SPHEQ', 'AL', 'ACD','LT','VCD','SPORTHR','READHR',
                                                'COMPHR', 'STUDYHR', 'TVHR', 'DIOPTERHR','MOMMY','DADMY',]])
```

```python
# Create new dataframe with standardized data
new_df2 = pd.DataFrame(scaled_data, columns=new_df.columns)
new_df2.head()
```

Part 2

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.90)
scaled_pca = pca.fit_transform(scaled_data)
```

```python
# Transform PCA data to a DataFrame
new_pca_df = pd.DataFrame(data=scaled_pca)
new_pca_df.head()
```

```python
# Initialize t-SNE model
from sklearn.manifold import TSNE
tsne = TSNE(learning_rate=45)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(new_pca_df)
```

```python
# The dataset has 2 columns
tsne_features.shape
```

```python
# Prepare to plot the dataset
new_pca_df['x'] = tsne_features[:,0]
new_pca_df['y'] = tsne_features[:,1]
```

```python
# Visualize the clusters
plt.scatter(new_pca_df['x'], new_pca_df['y'])
plt.show()
```

Part 3

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(new_pca_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for data')
plt.show()
print('The elbow of the plot is 5')
```

Recommmendation:I believe the patients would be able to be clustered. When looking at the scatter plot and elbow plot, I would recommend patients be placed into 5 clusters.
